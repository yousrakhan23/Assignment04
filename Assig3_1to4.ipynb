1. Project 1: Mad Libs Python Project - A word replacement game where users provide
random words that are inserted into a pre-defined story, creating humorous outcomes.
This project covers string manipulation and user input handling.

```python
# Mad Libs Generator in Google Colab

story = """
One day, {name} decided to go on a {adjective} adventure.
They packed their {noun} and set off to {verb} in the {place}.
Suddenly, a {animal} appeared and said, "{exclamation}!"
{name} was so {emotion} that they {verb_past_tense} all the way home.
"""

print("Welcome to Mad Libs! Please provide the following words:")

# Get user inputs
inputs = {
    "name": input("Enter a name: "),
    "adjective": input("Enter an adjective: "),
    "noun": input("Enter a noun: "),
    "verb": input("Enter a verb: "),
    "place": input("Enter a place: "),
    "animal": input("Enter an animal: "),
    "exclamation": input("Enter an exclamation (e.g., 'Wow!'): "),
    "emotion": input("Enter an emotion: "),
    "verb_past_tense": input("Enter a past-tense verb (e.g., 'ran'): ")
}

# Fill in the story
filled_story = story.format(**inputs)

# Display the result
print("\nYour Mad Libs Story:")
print(filled_story)
```

2. Project 2: Guess the Number Game (Computer) - In this version, the computer tries to
guess a number that the user is thinking of. The user provides feedback (higher/lower)
until the computer guesses correctly. This project introduces binary search concepts
and logical loops.

```python
def computer_guesses_fancy():
    from IPython.display import clear_output
    import time

    print("""\n
    â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–ˆâ–‘â–ˆâ–€â–„â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–€â–‘â–€â–ˆâ–€â–‘â–ˆâ–€â–ˆâ–‘â–ˆâ–‘â–ˆ
    â–‘â–ˆâ–‘â–‘â–‘â–ˆâ–‘â–ˆâ–‘â–ˆâ–€â–„â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–€â–‘â–‘â–ˆâ–‘â–‘â–ˆâ–‘â–ˆâ–‘â–ˆâ–‘â–ˆ
    â–‘â–€â–€â–€â–‘â–€â–€â–€â–‘â–€â–‘â–€â–‘â–€â–€â–€â–‘â–€â–‘â–‘â–‘â–€â–€â–€â–‘â–€â–‘â–€â–‘â–€â–€â–€
    """)
    time.sleep(1)

    print("Think of a secret number between 1-100.")
    print("I'll try to guess it in the fewest attempts possible!")
    time.sleep(1.5)

    print("\nRespond with:")
    print("â†’ 'h' if my guess is too HIGH")
    print("â†’ 'l' if my guess is too LOW")
    print("â†’ 'c' if my guess is CORRECT")
    time.sleep(2)

    low, high = 1, 100
    attempts = 0
    guess_history = []

    while True:
        clear_output(wait=True)
        guess = (low + high) // 2
        attempts += 1

        print("\n" + "="*40)
        print(f"ATTEMPT #{attempts}")
        print("="*40)

        print("\nMy guess is:", guess)
        print("\nPrevious guesses:", guess_history or "None yet")

        feedback = input("\nYour feedback? (h/l/c): ").lower()

        if feedback == 'c':
            print(f"\n{'ğŸ‰'*3} VICTORY! {'ğŸ‰'*3}")
            print(f"I guessed your number {guess} in just {attempts} attempts!")
            break
        elif feedback == 'h':
            high = guess - 1
            guess_history.append(f"{guess} (Too high)")
        elif feedback == 'l':
            low = guess + 1
            guess_history.append(f"{guess} (Too low)")
        else:
            print("âš ï¸ Invalid input! Use 'h', 'l', or 'c'")
            attempts -= 1
            time.sleep(1)
            continue

        if low > high:
            print("\n{'ğŸ¤”'*3} Hmm... something's wrong!")
            print("You must have changed your number!")
            break

        time.sleep(0.5)

# Start the enhanced game
computer_guesses_fancy()
```

3. Project 3: Guess the Number Game (User) - The user attempts to guess a randomly
generated number by the computer. The program provides hints (higher/lower) until
the correct number is guessed. This project focuses on loops, conditionals, and
random number generation.

```python
import random
from IPython.display import clear_output

def number_guessing_game():
    # Game setup
    print("""\n
    â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–ˆâ–‘â–ˆâ–€â–„â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–€â–‘â–€â–ˆâ–€â–‘â–ˆâ–€â–ˆâ–‘â–ˆâ–‘â–ˆ
    â–‘â–ˆâ–‘â–‘â–‘â–ˆâ–‘â–ˆâ–‘â–ˆâ–€â–„â–‘â–ˆâ–€â–€â–‘â–ˆâ–€â–€â–‘â–‘â–ˆâ–‘â–‘â–ˆâ–‘â–ˆâ–‘â–ˆâ–‘â–ˆ
    â–‘â–€â–€â–€â–‘â–€â–€â–€â–‘â–€â–‘â–€â–‘â–€â–€â–€â–‘â–€â–‘â–‘â–‘â–€â–€â–€â–‘â–€â–‘â–€â–‘â–€â–€â–€
    """)

    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")

    # Generate random number
    secret_number = random.randint(1, 100)
    attempts = 0
    guess_history = []

    while True:
        try:
            # Get user guess
            guess = int(input("\nTake a guess: "))
            attempts += 1

            # Check guess
            if guess < 1 or guess > 100:
                print("Please enter a number between 1-100")
                attempts -= 1
                continue

            guess_history.append(guess)

            if guess < secret_number:
                print("Too low! ğŸ“‰ Try a higher number.")
            elif guess > secret_number:
                print("Too high! ğŸ“ˆ Try a lower number.")
            else:
                print(f"\nğŸ‰ Congratulations! You guessed it in {attempts} attempts!")
                print(f"Your guesses: {guess_history}")
                break

        except ValueError:
            print("Please enter a valid number!")

    print("\nThanks for playing!")

# Start the game
number_guessing_game()
```
